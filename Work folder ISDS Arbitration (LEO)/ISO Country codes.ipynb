```python


import pandas as pd


import pycountry
```

```python


# Load CSV
df = pd.read_csv('ISDS_sector_binary_dataset.csv')

# Function to convert country name to ISO code
def get_iso_code(country_name):
    try:
        country = pycountry.countries.search_fuzzy(country_name)[0]
        return country.alpha_3  
        # return country.alpha_2  # For 2-letter code 
    except:
        return None  # Returns None if country not found

# Apply to your columns
df['respondent_state'] = df['respondent_state'].apply(get_iso_code)
df['investor_home_state'] = df['investor_home_state'].apply(get_iso_code)

# Save the result
df.to_csv('output_with_iso_codes.csv', index=False)

# Check which countries couldn't be matched
unmatched = df[df['respondent_state'].isna() | df['investor_home_state'].isna()]
print(f"Unmatched countries: {len(unmatched)}")
print(unmatched[['respondent_state', 'investor_home_state']])
```

```python
# Load CSV
df = pd.read_csv('ISDS_sector_binary_dataset.csv')

# Function to convert country name(s) to ISO code(s)
def get_iso_code(country_name):
    # Handle NaN/None values
    if pd.isna(country_name):
        return None
    
    # Check if there are multiple countries separated by semicolon
    if ';' in str(country_name):
        countries = [c.strip() for c in str(country_name).split(';')]
        iso_codes = []
        for country in countries:
            try:
                country_obj = pycountry.countries.search_fuzzy(country)[0]
                iso_codes.append(country_obj.alpha_3)
            except:
                iso_codes.append(None)  # Keep None for unmatched countries
        # Join back with semicolon, filtering out None values
        valid_codes = [code for code in iso_codes if code is not None]
        return '; '.join(valid_codes) if valid_codes else None
    else:
        # Single country (original logic)
        try:
            country = pycountry.countries.search_fuzzy(str(country_name))[0]
            return country.alpha_3  
            # return country.alpha_2  # For 2-letter code 
        except:
            return None  # Returns None if country not found

# Apply to your columns
df['respondent_state'] = df['respondent_state'].apply(get_iso_code)
df['investor_home_state'] = df['investor_home_state'].apply(get_iso_code)

# Save the result
df.to_csv('output_with_iso_codes.csv', index=False)

# Check which countries couldn't be matched
unmatched = df[df['respondent_state'].isna() | df['investor_home_state'].isna()]
print(f"Unmatched countries: {len(unmatched)}")
print(unmatched[['respondent_state', 'investor_home_state']])
```
